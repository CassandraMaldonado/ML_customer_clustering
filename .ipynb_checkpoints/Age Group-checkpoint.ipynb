```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
file_path = "shopping_behavior_updated.csv"
df = pd.read_csv(file_path)

# Display basic information and first few rows
df.info(), df.head()
```

```python
# Define age groups
bins = [18, 25, 35, 45, 55, 65, 100]  # Age ranges
labels = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]

# Create Age Group column
df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)

# Get the total count for each age group for normalization
age_group_counts = df["Age Group"].value_counts()

# Function to normalize counts by age group proportion
def normalize_by_age_group(column):
    return df.groupby(["Age Group", column]).size().unstack().div(age_group_counts, axis=0)

# 1. Normalized Product Category Preferences by Age Group
plt.figure(figsize=(12, 6))
normalize_by_age_group("Category").T.plot(kind="bar", figsize=(12, 6), edgecolor='black', alpha=0.7)
plt.xlabel("Category")
plt.ylabel("Proportion of Age Group")
plt.title("Normalized Product Category Preferences by Age Group")
plt.xticks(rotation=45)
plt.legend(title="Age Group")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Normalized Shopping Frequency by Age Group
plt.figure(figsize=(12, 6))
normalize_by_age_group("Frequency of Purchases").T.plot(kind="bar", figsize=(12, 6), edgecolor='black', alpha=0.7)
plt.xlabel("Frequency of Purchases")
plt.ylabel("Proportion of Age Group")
plt.title("Normalized Shopping Frequency by Age Group")
plt.xticks(rotation=45)
plt.legend(title="Age Group")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 3. Normalized Discount Usage by Age Group
plt.figure(figsize=(10, 6))
normalize_by_age_group("Discount Applied").T.plot(kind="bar", figsize=(10, 6), edgecolor='black', alpha=0.7)
plt.xlabel("Discount Applied")
plt.ylabel("Proportion of Age Group")
plt.title("Normalized Discount Usage by Age Group")
plt.legend(title="Age Group")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 4. Normalized Payment Method Preferences by Age Group
plt.figure(figsize=(12, 6))
normalize_by_age_group("Payment Method").T.plot(kind="bar", figsize=(12, 6), edgecolor='black', alpha=0.7)
plt.xlabel("Payment Method")
plt.ylabel("Proportion of Age Group")
plt.title("Normalized Payment Method Preferences by Age Group")
plt.xticks(rotation=45)
plt.legend(title="Age Group")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
